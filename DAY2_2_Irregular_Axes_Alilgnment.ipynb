```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
#plt.subplot2grid
fig = plt.figure(figsize = (7,7))
ax = plt.subplot2grid((1,1),(0,0),fig = fig)
```

```python
fig = plt.figure(figsize = (7,7))
ax1 = plt.subplot2grid((2,1),(0,0),fig = fig) # (2,1)-> shape of axes, #(0,0) index of axes
ax2 = plt.subplot2grid((2,1),(1,0),fig = fig)
```

```python
fig = plt.figure(figsize = (7,7),facecolor = 'linen')
ax1 = plt.subplot2grid((3,3),(0,0),colspan = 2, fig = fig)
ax2 = plt.subplot2grid((3,3),(1,0),colspan = 3, fig = fig)
```

```python
fig = plt.figure(figsize = (7,7),facecolor = 'linen')
ax1 = plt.subplot2grid((2,2),(0,0),colspan = 2,fig = fig)
ax2 = plt.subplot2grid((2,2),(1,0),fig = fig)
ax3 = plt.subplot2grid((2,2),(1,1),fig = fig)
```

```python
fig = plt.figure(figsize = (7,7),facecolor = "linen")

ax1 = plt.subplot2grid((3,3),(0,0),
                       colspan = 2,
                       fig = fig)

ax2 = plt.subplot2grid((3,3),(1,0),
                       colspan = 2,
                       fig = fig)

ax3 = plt.subplot2grid((3,3),(2,0),
                       colspan = 2,
                       fig = fig)

ax4 = plt.subplot2grid((3,3),(0,2),
                       rowspan = 3,
                       fig = fig)
```
