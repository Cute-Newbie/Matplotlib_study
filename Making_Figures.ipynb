```python
#1-1.Figures and Axes

# Creating figure : Creating space for graph or picture
figure = plt.figure(figsize = (7,7),
                    facecolor = 'linen')

# Add Subplots
ax = figure.add_subplot()
ax.plot([2,3,1])
ax.scatter([2,3,1],[2,3,4])

```

```python
# Add Subplots: More graphs on single figure
figure1 = plt.figure()
ax1 = figure1.add_subplot(3,1,1)
ax2 = figure1.add_subplot(3,1,2)
ax3 = figure1.add_subplot(3,1,3)
ax1.plot([2,5,10])
ax2.plot([10,5,2])
ax3.plot([2,3,4])
```

```python
figure2 = plt.figure()
ax1_1 = figure2.add_subplot(2,2,1)
ax1_2 = figure2.add_subplot(2,2,2)
ax2_1 = figure2.add_subplot(2,2,3)
ax2_2 = figure2.add_subplot(2,2,4)
```
